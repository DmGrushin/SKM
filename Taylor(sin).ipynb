## 1. Ряд Тейлора для синуса
Напишите функцию `mysin(x, n)`, вычисляющую сумму первых $n$
членов ряда Тейлора для функции $\sin x$:
$$
\sin x = x - \frac{x^3}{3!} + \frac{x^5}{5!} - \dots
$$
Функция должна выполнять $O(n)$ операций и не содержать питоновских циклов. 
Постройте графики получающихся функций для разных $n$.
Вариант этой функции `mysin(x)` должен вычислять столько слагаемых,
пока результат перестанет изменяться (он может содержать циклы).
Протестируйте функцию на значениях 
$x=\frac{\pi}{2},\,\frac{11\pi}{2},\,\frac{21\pi}{2},\,\frac{31\pi}{2}$.
Постройте графики зависимости точности от числа слагаемых $n$.
Какова точность результатов, полученных для таких $x$? 
Сколько членов потребовалось?
Каков максимальный по абсолютному значению член?
Попробуйте объяснить, почему результаты
неудовлетворительные.

```python
import math # Объявление библиотек.
import numpy as np
import matplotlib.pyplot as plt

def mysin(x, n):

    q = x # Объявление переменных.
    sm = 0
    last_sm = 0

    max_q = 0
    data_x = [] # Создание массивов, чтобы хранить данные о сумме рядов (для отрисовки графика).
    data_y = []

    
    i = 1
    flag = True
    while i < n and flag:
        q = q * (-1) * (x * x) / ((2 * i + 1) * (2 * i)) # Используя формулы разложения элементарной функции sin x, выведем уравнение.
        last_sm = sm # Запоминание последнего результата суммы.
        sm = sm + q # Прибавляем sm (сумма ряда) к q (член ряда).
        if last_sm == sm: # Проверка совпадения последнего результата с новым. Если совпадает, то происходит прекращение работы.
            pass
        if q>max_q: # Проверка: больше ли последний член, чем тот что хранится в переменной? Если больше, то записываем его в max_q.
            max_q = q
        print(sm)
        data_x.append(sm) # Добавляем в массивы наши "точки" , то есть сумму после каждого приавления.
        data_y.append(i) # Добавляем количество n в виде точек.
        i += 1
    
    print("Потребовалось ", i, "членов")

    plt.plot(data_y, data_x) # Отрисовка графика.
    plt.ylabel('Сумма первых n членов ряда') # Текстовые метки координатных осей.
    plt.xlabel('n')
    print("Максимальный по абсолютному значение член = ", max_q) # Выводим в консоль.
    plt.show()

# Вызовем нашу функцию с разными аргументами, предоставленной условием задачи: pi/2, 11*pi/2, 21*pi/2, 31*pi/2.
mysin(math.pi/2, 1000)
mysin((11*math.pi)/2, 25)
mysin((21*math.pi)/2, 25)
mysin((31*math.pi)/2, 25)
```

Вычисление суммы ряда с заданной точностью ε следует производить до тех пор, пока не выполнится условие |un| ≤ ε. В данном примере вычисление суммы ряда$$
y = \frac{x}{2} - \frac{x^2}{2^2} + \frac{x^3}{2^3} - \frac{x^4}{2^4} + \dots
$$необходимо производить до тех пор, пока очередной член ряда по абсолютной величине не будет меньше ε = 0.001. Обозначим n–й член ряда. 
В нашем случае мы производим вычисления до того момента пока последние два члена не совпадут (не будет изменений). 
Для того чтобы определить точность результатов, уберем проверку совпадений: ( if last_sm == sm ).

```python
import math # Объявление библиотек.
import numpy as np
import matplotlib.pyplot as plt

def mysin(x, n):

    q = x # Объявление переменных.
    sm = 0
    last_sm = 0

    max_q = 0
    data_x = [] # Создание массивов, чтобы хранить данные о сумме рядов (для отрисовки графика).
    data_y = []
    

    i = 1
    flag = True
    while i < n and flag:
        q = q * (-1) * (x * x) / ((2 * i + 1) * (2 * i)) # Используя формулы разложения элементарной функции sin x, выведем уравнение.
        last_sm = sm # Запоминание последнего результата суммы.
        sm = sm + q # Прибавляем sm (сумма ряда) к q (член ряда).
        if q>max_q: # Проверка: больше ли последний член, чем тот что хранится в переменной? Если больше, то записываем его в max_q.
            max_q = q
        print(sm)
        data_x.append(sm) # Добавляем в массивы наши "точки" , то есть сумму после каждого приавления.
        data_y.append(i) # Добавляем количество n в виде точек.
        i += 1
    
    print("Потребовалось ", i, "членов")

    plt.plot(data_y, data_x) # Отрисовка графика.
    plt.ylabel('Сумма первых n членов ряда') # Текстовые метки координатных осей.
    plt.xlabel('n')
    print("Максимальный по абсолютному значение член = ", max_q) # Выводим в консоль.
    plt.show()

# Вызовем нашу функцию с разными аргументами, предоставленной условием задачи: pi/2, 11*pi/2, 21*pi/2, 31*pi/2.
mysin(math.pi/2, 1000)
mysin((11*math.pi)/2, 2)
mysin((21*math.pi)/2, 5)
mysin((31*math.pi)/2, 5)
```

Убрав проверку совпадения видно, что дальнейший результат вовсе не меняется. Так что точность ограничивается способностью машины.
